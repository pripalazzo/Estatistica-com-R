```
library(pacman)

p_load(DataExplorer,readxl, tidymodels,vip)
```

```
df <- read.csv('Automobile_data_1.csv', 
                 na = "?", sep = ";")

View(df)
```

```
DataExplorer::plot_scatterplot(df, by = 'price')
```

```
DataExplorer::create_report(df, y = "price")
```

```
#plotando o gráfico Q-Q plot para visualizar a distribuição
qqnorm(df$price)
qqline(df$price)
```

```
df$MARCA <- NULL
split <- initial_split(df, prop = .8, strata = Preco)
df_treino <- training(split)
df_teste <- testing(split)



DataExplorer::create_report(df, y = 'Preco')

fit <- lm(Preco ~ . , data = na.omit(df_treino))

summary(fit)

fit_step <- step(fit)

summary(fit_step)

pred <- predict(fit, df_teste)

plot(x = df_teste$Preco , y = pred, pch = 16)
```
